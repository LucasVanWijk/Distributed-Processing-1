- PN	Laagste uurwaarde van de luchtdruk herleid tot zeeniveau (in 0.1 hPa)

## Opzetten KNeighbours
Er wordt een tabel aangemaakt met daarop alle start- en eindcoordinaten met een label start of niet

```python
df_start = df[['start_lat','start_lng','start_time']]
df_start.columns = ['lat','lng','timestamp']
df_start['is_start'] = True

df_end = df[['end_lat','end_lng', 'end_time']]
df_end.columns = ['lat','lng','timestamp']
df_end['is_start'] = False

df_coords = pd.concat([df_start,df_end])
df_coords
```

```python
X = df_coords[['lat','lng']]
y = df_coords['is_start']
```

```python
n = 25
clf = KNeighborsClassifier(n_neighbors=n, n_jobs=2, weights='distance')
clf.fit(X, y)
```

```python
def heatmap(area,df_coords,resolution=.0005, scatter=False, samples=0):
    X = df_coords[['lat','lng']]
    y = df_coords['is_start']
    x_min, x_max, y_min, y_max = area
    # Grid size
    h = resolution

    # Meshgrid init
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
    np.arange(y_min, y_max, h))

    # Colormap
    cmap_bold = ListedColormap(['lightgray', 'yellow'])
    cmap_light = ListedColormap(['black', 'orange'])

    # predict class using data and kNN classifier
    Z = clf.predict(np.c_[yy.ravel(), xx.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    fig = plt.figure(figsize=[10,10])
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)
    
    if scatter:
        if samples:
            s = df_coords.sample(samples)
            X = s[['lat','lng']]
            y = s['is_start']

        plt.scatter(X.lng, X.lat, c=y, cmap=cmap_bold, s=2)
        plt.xlim(xx.min(), xx.max())
        plt.ylim(yy.min(), yy.max())
    return fig
```

## Omgeving verdelen in vierkante hokjes en het fiets overschot/tekort weergeven
Hieronder verdeel ik de omgeving in squares en plot ik met een heatmap het overschot/tekort

```python
def heatmap(test_df, area, resolution, timeframe=None):
    if timeframe != None:
        start_date, end_date = timeframe
        mask = (test_df['timestamp'] > start_date) & (test_df['timestamp'] <= end_date)
        tmp_df = test_df.loc[mask]
        title = "Heatmap " + str(timeframe[0] + " - " + str(timeframe[1]))
    else:
        tmp_df = test_df
        title = "Heatmap (Gehele dataset)"

    min_lat, max_lat = area[2], area[3]
    min_lng, max_lng = area[0], area[1]
    scope_lat = abs(max_lat - min_lat) / resolution
    scope_lng = abs(max_lng - min_lng) / resolution
    squares = np.zeros([resolution,resolution])


    for ix, x in enumerate(squares):
        square_lat_min, square_lat_max = min_lat + scope_lat * ix, max_lat + scope_lat * ix
        for yx, y in enumerate(x):
            square_lng_min, square_lng_max = min_lng + scope_lng * yx, max_lng + scope_lng * yx

            square_rides = tmp_df[tmp_df.lat>=square_lat_min]
            square_rides = square_rides[square_rides.lat<=square_lat_max]
            square_rides = square_rides[square_rides.lng>=square_lng_min]
            square_rides = square_rides[square_rides.lng<=square_lng_max]

            start_count  = square_rides[square_rides.is_start==True].count()[0]
            end_count = square_rides[square_rides.is_start==False].count()[0]

            diff = start_count - end_count
            squares[ix,yx] = diff


    xx, yy = np.meshgrid(np.arange(min_lat, max_lat, scope_lat),
    np.arange(min_lng, max_lng, scope_lng))
    fig = plt.figure(figsize=[10,7])
    f = plt.pcolormesh(yy, xx, squares, cmap='coolwarm')
    fig.colorbar(f)
    plt.title(title)
    plt.show()
```

### Opvangen pieken


In eerste instantie is het belangrijk om te achterhalen hoelang een rit gemiddeld duurt. Als wij dan zouden weten hoeveel fietsen op een moment in gebruik zijn hoe lang het duurt voordat die fiets weer vrij komt.

```python
#Makes a copy of the dataframe so that we don't contaminate the original.
occupation_df = df.copy()

#Converts time to datetime format form string format
occupation_df['start_time'] = pd.to_datetime(occupation_df['start_time'], format="%Y-%m-%d")
occupation_df['end_time'] = pd.to_datetime(occupation_df['end_time'], format="%Y-%m-%d")
occupation_df['diff'] = occupation_df['end_time'] - occupation_df['start_time']

def temp(x):
    return np.count_nonzero(occupation_df['diff'].dt.total_seconds() < x)

print("Aantal ritten onder de 1 uur      {} van de {} dat is {} % van totaal".format(temp(3600), len(occupation_df['diff']), round(temp(3600) / len(occupation_df['diff']) * 100, 2)))
print("Aantal ritten onder de 30 minuten {} van de {} dat is {} % van totaal".format(temp(1800), len(occupation_df['diff']), round(temp(1800) / len(occupation_df['diff']) * 100, 2)))
print("Aantal ritten onder de 15 munten  {} van de {} dat is {} % van totaal".format(temp(900), len(occupation_df['diff']), round(temp(900) / len(occupation_df['diff']) * 100, 2)))
print("Aantal ritten onder de 5 minuten   {} van de {} dat is {} % van totaal".format(temp(300), len(occupation_df['diff']), round(temp(300) / len(occupation_df['diff']) * 100, 2)))
```

Hieruit er van uitgaande dat de data klopt kunnen we concluderen dat als een fiets in gebruik is wij in 85% van de gevallen kunnen stellen dat die fiets binnen 15 minuten weer beschikbaar is. De fietsen worden dus echt voor korten ritjes gerbuikt. 

### Wanneer worden de fietsen gebruikt

Nu is het belangrijk te achterhalen wanneer de fietsen in gebruik zijn. Wij zullen dit proberen te achterhalen door voor elke minuut in de dag te kijken hoeveel fietsen in gebruik zijn. Wij zullen eerst kijken hoe dit er uit ziet voor alle dagen in de data set. 

```python
#Makes a copy of the dataframe so that we don't contaminate the original.
occupation_df = df.copy()

#Splits the date and the time into two different columns.
#This cannot be done in the cell above because here they need to be time
#for the time comperisons en above they need to be in datetime format
occupation_df['start_date'] = pd.to_datetime(df['start_time']).dt.date
occupation_df['start_time'] = pd.to_datetime(df['start_time']).dt.time
occupation_df['end_date'] = pd.to_datetime(df['end_time']).dt.date
occupation_df['end_time'] = pd.to_datetime(df['end_time']).dt.time

def calc_ussage(start_time,end_time):
    #Checks how many bike have been rented at every minute of the day
    list_of_mask = []
    colum_names = ["{}:{}".format(minute // 60, minute % 60) for minute in range(0,1440)]
    for i in colum_names:
        time = pd.to_datetime(i, format='%H:%M').time()
        a = (time > start_time) & (time < end_time)
        list_of_mask.append(a)

    oc_df = pd.concat(list_of_mask, axis=1)
    oc_df.columns = colum_names
    return oc_df.sum()

def plot_ussage(title_string,ussage):
    ussage.plot(figsize=(30,15))
    #plt.xticks(pd.Series.tolist(ussage), colum_names[::15], rotation='vertical')
    plt.yticks(fontsize=20)
    plt.title(title_string, fontsize=50)
    plt.show()

ussage = calc_ussage(occupation_df['start_time'],occupation_df['end_time'])
plot_ussage("Total daily occupation",ussage)

```

Wij zien hier een grafiek die op de Y as aangeeft hoe vaak fietensen over de helen dataset op het zelfde moment in gebruik waren op een bepaalde minuut. Het specifieke nummer is voor ons niet zo relevant maar vooral hoe hoog deze  is relatief aan de ander punten. Dit geeft ons dan aan wanneer de pieken plaats vinden


Hier zien wij dat de grootste piek van tegelijk in gebruik gemiddeld plaatst vind tussen 12:00 en 13:00 met een aantal kleinere pieken rond 15:00 17:00 en 09:00. Wat opvalt is dat de piek om 09:00 niet zo groot is als die van 17:00. Dit zou kunnen komen door dat mensen meer rond de zelfde tijd weggaan dan ze aankomen


```python
workdays = occupation_df.copy()
workdays['weekday'] = pd.to_datetime(df['start_time'], format="%Y-%m-%d").dt.weekday.astype('int')
workdays['date'] = pd.to_datetime(df['date'], format="%Y-%m-%d")
workdays = workdays.loc[(workdays['weekday'] < 5)]

ussage = calc_ussage(workdays['start_time'],workdays['end_time'])
plot_ussage("work day occupation",ussage)
```

Deze grafiek laat redelijk het zelfde zien als de vorig met ongeveer de zelfde pieken. Nu gaan we kijken naar de situatie van de afgelopen drie maanden die zitten natuurlijk dichter bij de huidige situatie dan de eerdere data omdat sind die tijd het gebruik redelijk is toegenomen.

Hier zien wij dat de pieken op de dagen individueel ook erg verschillen. We kunnen dus wel bepalen wanneer de pieken ongeveer zijn maar niet precies. Ook kunnen we niet exact stellen hoeveel fietsen er nodig zullen zijn. Wel zien wij dat er hier over het algemeen tussen de 10 en 17 fietsen nodig zijn. Toch willen we iets specifieker zijn du

```python
def get_max_same_time_occ_for_list_days(days):
    #Loops over all the workdays in the last three months
    #checks how many bikes are rented at the same time and adds the higest amount over the day to a list
    i = 0
    presentage_shown = []
    max_occupation_per_day = []
    for date in days:  
        current_day = workdays.loc[workdays['start_date'] == pd.to_datetime(date)]
        ussage = calc_ussage(current_day['start_time'],current_day['end_time'])
        max_occupation_per_day.append(ussage.max())

        i += 1
        presentage_done =  round((i / len(days) *100), 0)
        if presentage_done not in presentage_shown and presentage_done % 5 == 0:
            print(str(presentage_done) + "%")
            presentage_shown.append(presentage_done)
        
    return max_occupation_per_day

days = workdays_last_three_months['start_date'].unique()
max_occupation_per_day = pd.Series(np.array(get_max_same_time_occ_for_list_days(days)))
print(max_occupation_per_day.describe())
max_occupation_per_day.plot(figsize=(30,15))
    
```
